## Practica 1 Metodo del gradiente

```python
# Importamos librerias y resolvemos

import numpy as np
import matplotlib.pyplot as plt

# Definir la función y su derivada
def f(x):
    return 2 * x**2 * np.cos(x) - 5 * x

def df(x):
    # Derivada de la función f(x)
    return 4 * x * np.cos(x) - 2 * x**2 * np.sin(x) - 5

# Implementación del algoritmo de descenso del gradiente
def gradient_descent(learning_rate, max_iters, tol):
    # Inicializar un punto aleatorio x0
    x = -1
    history = [x]
    
    for i in range(max_iters):
        grad = df(x)
        new_x = x - learning_rate * grad  # Actualizar x en dirección opuesta al gradiente
        history.append(new_x)
        
        # Parar si el cambio es menor que la tolerancia
        if np.abs(new_x - x) < tol:
            break
        
        x = new_x
    
    return x, history

# Parámetros del algoritmo
learning_rate = 0.01
max_iters = 1000
tol = 1e-6

# Ejecutar el descenso del gradiente
optimal_x, history = gradient_descent(learning_rate, max_iters, tol)

# Mostrar el resultado
print(f"Valor óptimo de x: {optimal_x}")
print(f"Valor óptimo de f(x): {f(optimal_x)}")
```

```python
## Graficamos
# Graficar la función y la convergencia
x_vals = np.linspace(-5, 5, 400)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label="f(x)")
plt.scatter(history, [f(x) for x in history], color='red', label="Descent path")
plt.title("Descenso del Gradiente")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
