```python
## Practica 8 La busqeuda local es una tecnica utilizada para resolver problemas de opitmizacion donde la
##solucion optima se busca moviendose inteativamente entre soluciones vecinas en un espacio de busqueda.

## Hillclimbing es un de los algoritomos mas simples de busqueda local. Comienza con una solucion inicial
## y realiza moviementos hacia soluciones vecinas que mejoran la funcion objetivo. 

## Funcion: -x^2 + 10

# Librerias 
import random
import matplotlib.pyplot as plt
import numpy as np

#Funcion objetivo 
def objective_function(x):
    return -x**2 + 10

#Funcion para definir el vecindario
def vecindario(x,step_size=0.1):
    return [x - step_size, x + step_size]

#Funcion hillclimbing
def hill_climbing(start, step_size = 0.1, max_iterations = 1000):
  current_solution = start
  current_value = objective_function(current_solution)
  history = [(current_solution, current_value)]

  for iteration in range(max_iterations):
      #Generar vecinos
      neighbors = vecindario(current_solution, step_size)

      #Evaluar los vecinos y seleccionar el mejor
      next_solution = max(neighbors, key=objective_function)
      next_value = objective_function(next_solution)

      #Alamcenar la solucion y valor
      history.append((next_solution, next_value))

      print(f"Iteration {iteration}: Current Value = {current_value}, Next Value = {next_value}")

      # Si el vecino es mejor, actualizar la solucion actual
      if next_value > current_value:
          current_solution = next_solution
          current_value = next_value
      else:
          # Termina si no puede mejorar más
          print("Se ha alcanzado un maximo local.")
          break

  return history

# Definir grafica
def plot(history):
    x = np.linspace(-5, 5, 400)
    y = objective_function(x)

    # Extraemos soluciones
    solutions= [sol[0] for sol in history]
    values = [sol[1] for sol in history]
    # Graficar
    plt.figure(figsize=(12, 6))
    plt.plot(x, y, label="f(x) = -x^2 + 10")
    plt.scatter(solutions, values, color='red', label="Hill Climbing")
    plt.plot(solutions, values, color='red', linestyle='--')
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.title("Hill Climbing")
    plt.legend()
    plt.grid(True)
    plt.show()

initial_solution = random.uniform(-5, 5)
history = hill_climbing(initial_solution)
plot(history)
```
